```python
# FINDING AND REMOVING OUTLIERS

import pandas as pd
import numpy as np
%matplotlib inline
import seaborn as sbn
# Load the file
df = pd.read_csv("d:/machine learning/german_credit_data.csv")
```

```python
# Drawing the boxplot of age
sbn.boxplot(df['Age'])
```

```python
# Computing the intraquartile range
Q1 = df["Age"].quantile(0.25)
Q3 = df["Age"].quantile(0.75)
IQR = Q3 - Q1
print("The intraquartile range is", IQR)
```

```python
# We find the upper fence and lower fence by adding the following code, and print all the data above the upper fence 
# and below the lower fence
Lower_Fence = Q1 - (1.5 * IQR)
Upper_Fence = Q3 + (1.5 * IQR)
print("The lower fence is", Lower_Fence)
print("The upper fence is", Upper_Fence)
```

```python
# To print all the data above the upper fence and below the lower fence
df[((df["Age"] < Lower_Fence) |(df["Age"] > Upper_Fence))]
```
